```python
from datetime import datetime, date, time
def parse_date(date):
    date = str(date)
    if (date == 'none')or(date =='unknown'):
        return NaT
    elif (len(date)==4):
        return datetime.strptime(date, "%Y")
    elif (len(date) >= 8):
        return datetime.strptime(date, "%m/%d/%Y")
    
```

```python
y=df_regression['GDP_per_capita']
x=df_regression['Life Expectancy']

Y = model.beta.intercept + model.beta.x*x
plt.xlabel("Life Expectancy")
plt.ylabel("GDP per capita")
plot(x,Y,'r-',x,y,'o')
```

###### The equation for predicting the value of average income(GDP per capita) in terms of Life expectancy is <br/> GDP_per_capita  = 1380.564445 * (Life Expectancy) - 81316.038641

```python
df_regression = df_regression.set_index('Country Name')
countries_above_forecast = []
countries_below_forecast = []
for index, row in df_regression.iterrows():
    Z = row['GDP_per_capita'] - model.beta.intercept - model.beta.x*row['Life Expectancy']
    if Z > 0:
        countries_above_forecast.append(index)
    elif Z < 0:
        countries_below_forecast.append(index)
```

###Custom Ranking is calculated as<br/>
######Over all rank = min{rank(life-expectancy) + rank(literacy rate) + rank(GDP_per_capita) + rank(net migration rate) + rank(internet_user_ratio) + rank(population density)}

```python
def calculateRank(df):
    Countries = {}
    dataRows = df.to_dict()
    for key, value in dataRows.iteritems():
        ranksum = 0
        for variable, rank in value.iteritems():
            ranksum = ranksum + rank
        Countries[key] = ranksum
    return Countries
```

```python
countries_custom_rank = calculateRank(df_rank.T)
sorted_x = sorted(countries_custom_rank.iteritems(), key=operator.itemgetter(1))
countries_rankwise=[]
for i in sorted_x:
    countries_rankwise.append(i[0])
```

```python
df_similarity = df_names.drop(['Latitude of Capital','Longitude of Capital','Date of Founding/Independence','Country Code','Government Type','Capital City'], 1)

distanceMatrix = {}
for index1, row1 in df_similarity.iterrows():
    distanceMatrix[index1] = {}
    for index2, row2 in df_similarity.iterrows():
        distanceMatrix[index1][index2]=sp.spatial.distance.cosine(row1,row2)
        
```

```python
df_distance = DataFrame.from_dict(distanceMatrix)
```

```python
def nearestandfarthest(df):
    df_farthest_nearest = {}
    dictionary= df.to_dict()
    for key, value in dictionary.iteritems():
        value.pop(key, None)
        max_distance = -1000000000
        min_distance = 1000000000
        maxminDict = {}
        for variable, varValue in value.iteritems():
            if varValue < min_distance :
                maxminDict['strongest'] = variable
                min_distance = varValue
            if varValue > max_distance :
                maxminDict['weakest'] = variable
                max_distance = varValue
        df_farthest_nearest[key]=maxminDict
    return df_farthest_nearest
```
